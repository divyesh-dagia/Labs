```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data1 = np.loadtxt("noload1")
data2 = np.loadtxt("noload2")
data3 = np.loadtxt("noload3")
noload = np.vstack((data1,data2,data3))
x = noload[:,1]
y = noload[:,2]
plt.scatter(noload[:,0],noload[:,1], label= "pressure")
plt.scatter(noload[:,0],noload[:,2], label = "volume")
plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0.120,color='b',linestyle="-")
plt.axvline(x=0.360,color='b',linestyle="-")
print("period= ", (0.36-0.12))
plt.legend()
```

```python
def scatter_plot_area(x,y,bins):
    plt.scatter(x,y)
    min_x = min(x)
    max_x = max(x)
    area = 0
    x_step = (max_x-min_x)/bins
    for l in range(bins):
        x1 = min_x + x_step*l
        x2 = min_x + (l+1)*x_step
        y_list = []
        for q in range(len(x)):
            if x1<=x[q] < x2:
                y_list.append(y[q])
    
        y_mean = sum(y_list)/len(y_list)
        yhigh = [i for i in y_list if i >= y_mean] 
        ylow =  [i for i in y_list if i < y_mean]
    
        yhigh_mean = sum(yhigh)/len(yhigh)
        ylow_mean = sum(ylow)/len(ylow)
    
        #calculate area of rectangle and add it to running average 
        area = area + ((yhigh_mean - ylow_mean) * x_step)
    
    
        #Plot the rectanlge bin
        plt.plot([x1, x1], [yhigh_mean, ylow_mean], 'k-', lw=1)
        plt.plot([x2, x2], [yhigh_mean, ylow_mean], 'k-', lw=1)
        plt.plot([x1, x2], [yhigh_mean, yhigh_mean], 'k-', lw=1)
        plt.plot([x1, x2], [ylow_mean, ylow_mean], 'k-', lw=1)
    
    print("area = ", area)
```

```python
scatter_plot_area(x,y,25)
```

```python
def my_plot(x,y):
    def linearfunc(x,intercept,slope):
        y = x*slope+intercept
        return y
    fig = plt.figure()
    a_fit, cov = curve_fit(linearfunc,x,y)
    xx = np.arange(np.min(x),np.max(x),1E-6)
    fitfcn = np.polynomial.Polynomial(a_fit)
    slope = a_fit[1], np.sqrt(np.diag(cov)[1])
    intercept = a_fit[0],np.sqrt(np.diag(cov)[0])
    plt.plot(xx,fitfcn(xx), "k-")
    plt.scatter(x,y)
    plt.xlabel("1/\u03BB")
    plt.ylabel("ln(I\u03BB\u2075)")
    plt.title("")
    slope = a_fit[1], np.sqrt(np.diag(cov)[1])
    intercept = a_fit[0],np.sqrt(np.diag(cov)[0])
    #plt.axis((np.min(x[start:stop]),np.max(x[start:stop]),np.min(y[start:stop]),np.max(y[start:stop])))
    return slope,intercept, fig
    
```
